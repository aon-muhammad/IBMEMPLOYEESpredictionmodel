```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X=final.drop("Attrition",axis=1)
y=final["Attrition"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import (
    RandomForestClassifier,
    AdaBoostClassifier,
    GradientBoostingClassifier,
    BaggingClassifier,
    StackingClassifier
)
from sklearn.naive_bayes import GaussianNB

from sklearn.model_selection import (
    GridSearchCV,
    RandomizedSearchCV,
    cross_val_score,
    StratifiedKFold
)
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report,
    roc_auc_score,
    roc_curve
)
base_models = [
    ('rf', RandomForestClassifier(n_estimators=10, random_state=42)),
    ('svr', SVC(probability=True, random_state=42))
]
```

```python
models={
    "LogisticRegression":LogisticRegression(),
    "KNeighborsClassifier":KNeighborsClassifier(),
    "SVC":SVC(probability=True),
    "DecisionTreeClassifier":DecisionTreeClassifier(),
    "RandomForestClassifier":RandomForestClassifier(),
    "AdaBoostClassifier":AdaBoostClassifier(),
    "GradientBoostingClassifier":GradientBoostingClassifier(),
    "BaggingClassifier":BaggingClassifier(),
    "Naive Bayes":GaussianNB(),
    "StackingClassifier":StackingClassifier(
    estimators=base_models, 
    final_estimator=LogisticRegression()),
    "XGBoost":XGBClassifier(use_label_encoder=False,eval_matric='logloss')
}
score2=[]
```

```python
final.head()
```

```python
final.columns
```

```python
numeric=['Age','Education', 'HourlyRate', 'JobInvolvement','JobLevel', 'JobSatisfaction', 'MonthlyRate', 'NumCompaniesWorked','PercentSalaryHike', 'PerformanceRating',
       'RelationshipSatisfaction','StockOptionLevel', 'TotalWorkingYears','TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany',
       'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']
X_train[numeric]=scaler.fit_transform(X_train[numeric])
X_test[numeric]=scaler.transform(X_test[numeric])
```

```python
for i,(name,mod) in enumerate(models.items(),start=0):
    mod.fit(X_train,y_train)
    y_prediction=mod.predict(X_test)
    acc=accuracy_score(y_test,y_prediction)
    f1=f1_score(y_test,y_prediction)
    y_pred_prob = mod.predict_proba(X_test)[:, 1]
    auc = roc_auc_score(y_test, y_pred_prob)
    score2.append({"Name:":name,"Accuracy:":acc,"F1_Score:":f1,"AUC":auc})
```

```python
score2
scores_df = pd.DataFrame(score2)
print(scores_df)
```

```python
tuning_models = {
    "SVC": SVC(probability=True),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "GradientBoosting": GradientBoostingClassifier(),
}

# 2. Define the hyperparameter search spaces
param_grids = {
    "SVC": {
        'C': [0.1, 1, 10, 100],
        'gamma': ['scale', 'auto', 0.01, 0.1],
        'kernel': ['rbf', 'linear'],
        'class_weight': ['balanced']
    },
    "XGBoost": {
        'n_estimators': [50, 100, 200],
        'max_depth': [3, 5, 7, 9],
        'learning_rate': [0.01, 0.1, 0.2],
        'subsample': [0.7, 0.8, 1.0]
    },
    "LogisticRegression": {
        'C': [0.01, 0.1, 1, 10, 100],
        'penalty': ['l2'],
        'class_weight': ['balanced']
    },
    "GradientBoosting": {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.01, 0.05, 0.1],
        'max_depth': [3, 4, 5]
    },
}
```

```python
tunedresults=[]
for name,mod in tuning_models.items():
    classifier=RandomizedSearchCV(estimator=mod,param_distributions=param_grids[name],cv=5,scoring="f1", n_iter=20,
    random_state=42,
    n_jobs=-1)
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    tunedresults.append({
    "Model": name,
    "Best Params": classifier.best_params_,
    "Best CV Score": classifier.best_score_,
    "Accuracy": accuracy_score(y_test, y_pred),
    "F1": f1_score(y_test, y_pred),
    "AUC": roc_auc_score(
        y_test,
        classifier.predict_proba(X_test)[:, 1]
    )
    })

```

```python
result=pd.DataFrame(tunedresults)
result
model_svc = classifier.best_estimator_
svc_pred = model_svc.predict(X_test)
print(f1_score(y_test,svc_pred))
```

```python

# 1. Handle imbalance explicitly
scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()  # ~5.2

bestest = XGBClassifier(
    n_estimators=200,
    max_depth=3,
    learning_rate=0.05,        # lower, not 0.2
    subsample=0.7,
    colsample_bytree=0.7,
    min_child_weight=3,        # regularization: requires more samples per leaf
    reg_alpha=0.5,              # L1
    reg_lambda=1.0,             # L2
    scale_pos_weight=scale_pos_weight,
    eval_metric="logloss",
    early_stopping_rounds=20,
    random_state=42
)

bestest.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],
    verbose=False
)
```

```python
pred=bestest.predict(X_test)
print(accuracy_score(y_test,pred))
print(f1_score(y_test,pred))
```
